# 10.5 Analysis of Stochastic Processes

*"behavior that is not predictable forever" (non-periodic), but "variability is always there" (not temporally localized)*

### 10.5.1 Autocorrelation and Structure Functions

** Correlation Function:**

$ CF(\Delta t) = \frac{\lim_{T -> \infty} f(t) g(t+\Delta t)}{\sigma_f \sigma_g} $

Note that $CF(0) = 1$

*"If one time series is produced from another one, by simply shifting the time axis by $t_{\rm lag}$, the CF has a preak at $\Delta t = t_{\rm lag}$"*

*Examples:* I imagine reverberation mapping, but I know for sure it has applications to X-ray scattering

** Autocorrelation Function:**

$ f(t) = g(t) = y(t) $

$ ACF(\Delta t) = \frac{\lim_{T -> \infty} y(t) y(t+\Delta t)}{\sigma_y^2} $

*For processes that "retain memory" for a characteristic time $\tau$, ACF vanishes for $\Delta t >> \tau$*

*The ACF and PSD are "Fourier pairs"*

**Question:** I am trying to think of an example of an astrophysical process like this

** Structure Function **

$ SF(\Delta t) = SF_\infty [1 - ACF(\Delta t)]^{1/2}$

$SF_\infty$ is the "standard deviation of the time series evaluated over an infinitely large time interval

By definition $SF(0)=0$

Note also when $SF \propto t^{\alpha}$, the $PSD \propto (1/f)^{1+2\alpha}$

** $1/f$ and $1/f^2$**

+ "White noise" vs "pink noise" vs "red noise"

+ $1/f^2$ : Brownian motion, i.e. random walk

### 10.5.3: Autoregressive Models

*"provide a good general description of processes that reatin memory of previous states (but are not periodic)"*

Ex: Random walk, where each new value adds noise to the preceding value

$ y_i = y_{i-1} + e_i $

OR depending on $K$ preceding walks:

$ y_i = \Sigma_{j=1}^K a_j y_{i-j} + e_i $

ACF for AR(K) is non-zero for all lags, but does decay

*Example:* Distinguish a **random walk** ($a_1=1$) from **pure noise** ($a_1 = 0$)

### 10.5.4: Damped Random Walk Model

*"married-drunkard's walk"*

$ACF(t) = \exp(-t/\tau)$

PSD is:

+ $1/f^2$ at high frequencies 
+ flatter at low frequencies ($f << 2\pi / \tau$)

*Example:* Optical continuum variability of quasars
