A key difference between astronomical (and physical) time-series data processing and analysis and the use in most of the rest of the ML / Stats world is the focus on prediction. We are interested more in the underlying model, others are interested in prediction (aka getting rich). The is essentially a difference between fitting and regression.

### Key point of 10.2.1 : You can just fit time series data like you would any data set: see the entire rest of the book :)

A key thing to keep in mind throughout these sections is that any kind of filtering in Fourier or wavelet space is getting rid of some of the data. There are basically only three good reasons to do this *(see above comment)*:

-- **Data Compresesion** This is rarely used in the astronomical or astrophysical context, but it is well known that natural signals typically only take up a small part of wavelet space, and that periodic signals are also often over sampled, such that we can use only the first few Fourier modes

-- **Visual Inspection / Hypothesis Generation** It is often very hard to "see" a signal in the noise, and these kinds of methods often make themmore visible. This is true in image data as well, where various Fourier and Wavelet filetering modes are critical to teasing features out of the data. 

-- **Matched Fitering** Often we want to highlight a particular signal, at a particular frequency. By running a matched filter (which is to say filtering in Fourier or Wavelet or some other space) we can highlight that signal, and then use some other methods (thresholding etc) to select those features of interest.

####10.2.2: Fourier Decomposition

```python
# Author: Jake VanderPlas
# License: BSD
#   The figure produced by this code is published in the textbook
#   "Statistics, Data Mining, and Machine Learning in Astronomy" (2013)
#   For more information, see http://astroML.github.com
#   To report a bug or issue, use the following forum:
#    https://groups.google.com/forum/#!forum/astroml-general
import numpy as np
from matplotlib import pyplot as plt

from astroML.datasets import fetch_rrlyrae_templates

#----------------------------------------------------------------------
# This function adjusts matplotlib settings for a uniform feel in the textbook.
# Note that with usetex=True, fonts are rendered with LaTeX.  This may
# result in an error if LaTeX is not installed on your system.  In that case,
# you can set usetex to False.
from astroML.plotting import setup_text_plots
setup_text_plots(fontsize=8, usetex=False)

#------------------------------------------------------------
# Load the RR Lyrae template
templates = fetch_rrlyrae_templates()
x, y = templates['115r'].T

#------------------------------------------------------------
# Plot the results
fig = plt.figure(figsize=(5, 5))
fig.subplots_adjust(hspace=0)

kvals = [1, 3, 8]
subplots = [411, 412, 413]

for (k, subplot) in zip(kvals, subplots):
    ax = fig.add_subplot(subplot)

    # Use FFT to fit a truncated Fourier series
    y_fft = np.fft.fft(y)
    y_fft[k + 1:-k] = 0
    y_fit = np.fft.ifft(y_fft).real

    # plot the true value and the k-term reconstruction
    ax.plot(np.concatenate([x, 1 + x]),
            np.concatenate([y, y]), '--k', lw=2)
    ax.plot(np.concatenate([x, 1 + x]),
            np.concatenate([y_fit, y_fit]), color='gray')

    label = "%i mode" % k
    if k > 1:
        label += 's'

    ax.text(0.02, 0.1, label, ha='left', va='bottom',
            transform=ax.transAxes)

    if subplot == subplots[-1]:
        ax.set_xlabel('phase')
    else:
        ax.xaxis.set_major_formatter(plt.NullFormatter())

    if subplot == subplots[1]:
        ax.set_ylabel('amplitude')
    ax.yaxis.set_major_formatter(plt.NullFormatter())

    ax.set_xlim(0, 2)
    ax.set_ylim(1.1, -0.1)

ax = fig.add_subplot(414)
ax.set_yscale('log')

ax.plot(np.sqrt(y_fft[0:10].real**2 + y_fft[0:10].imag**2), lw=2)
plt.show()
```

This Fourier decomposition is equivalent to stating that most of the power exsists in the lowest 8 Fourier modes available to us, which are the ones represented in the 8 smallest and 8 highest values in the FFT

####DFTs/Sampling

One of the key concepts here is Nyquist/Shannon sampling, the idea that given a bandlimited signal one has all the information with a certain time sampling. There is a whole lot of math I am not going to go into here; if you want an excellent read on the DFT *(with perhaps a bit more IDL than you probably want)* check out http://astro.berkeley.edu/~heiles/handouts/fourierc.pdf -- no one beats old-school radio astronomers in deep understanding of Fouier space.
```python
import numpy as np
from scipy.stats import norm
from IPython.html import widgets
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-10,10,20)
x2 = np.linspace(-10,10,200)
def special_sin(a):
    y = np.sin(a*x)
    y2 = np.sin(a*x2)
    plt.figure(figsize=(6,6))
    plt.plot(x, y, marker=None,linestyle='-')
    plt.plot(x2, y2, marker=None,linestyle='-')
    plt.xlim(-10,10)
    plt.show()
```

####Wavelets

Wavelets are used by most of the community as a compression algorithm, based on the fact that the concentrate the information in an image into a small part of wavelet space, and the remaining wavelets can be thrown away. JPEG 2000 (and a number of still and video formats) use wavelets for compression. There are lots and lots of wavelets (Morlet, DoG, Haar) that can decompose 1D (and higher) data. 

In our context they are much more often used as either a) a matched filter or b) an way to highlight certain features in images (or signal streams).

####Low Pass Filters

Wiener filtering is a form of low-pass filtering, designed to maximally keep the signal and reduce the noise in a MMSE sense. The implementation in AstroML assumes that the signal is a Gaussian in Fourier space and the noise is white. There are many other implementations of Weiner filtering, but I thought i'd just show you this one, as it comes up a lot of in astronomical image processing.

```python
# Author: Jake VanderPlas
# License: BSD
#   The figure produced by this code is published in the textbook
#   "Statistics, Data Mining, and Machine Learning in Astronomy" (2013)
#   For more information, see http://astroML.github.com
#   To report a bug or issue, use the following forum:
#    https://groups.google.com/forum/#!forum/astroml-general
import numpy as np
from matplotlib import pyplot as plt

from scipy import optimize, fftpack
from astroML.filters import savitzky_golay, wiener_filter

#----------------------------------------------------------------------
# This function adjusts matplotlib settings for a uniform feel in the textbook.
# Note that with usetex=True, fonts are rendered with LaTeX.  This may
# result in an error if LaTeX is not installed on your system.  In that case,
# you can set usetex to False.
from astroML.plotting import setup_text_plots
setup_text_plots(fontsize=8, usetex=False)

#------------------------------------------------------------
# Create the noisy data
np.random.seed(5)
N = 2000
dt = 0.05

t = dt * np.arange(N)
h = np.exp(-0.5 * ((t - 20.) / 1.0) ** 2)
hN = h + np.random.normal(0, 0.5, size=h.shape)

Df = 1. / N / dt
f = fftpack.ifftshift(Df * (np.arange(N) - N / 2))
HN = fftpack.fft(hN)

#------------------------------------------------------------
# Set up the Wiener filter:
#  fit a model to the PSD consisting of the sum of a
#  gaussian and white noise
h_smooth, PSD, P_S, P_N, Phi = wiener_filter(t, hN, return_PSDs=True)

#------------------------------------------------------------
# Use the Savitzky-Golay filter to filter the values
h_sg = savitzky_golay(hN, window_size=201, order=4, use_fft=False)

#------------------------------------------------------------
# Plot the results
N = len(t)
Df = 1. / N / (t[1] - t[0])
f = fftpack.ifftshift(Df * (np.arange(N) - N / 2))
HN = fftpack.fft(hN)

fig = plt.figure(figsize=(5, 3.75))
fig.subplots_adjust(wspace=0.05, hspace=0.25,
                    bottom=0.1, top=0.95,
                    left=0.12, right=0.95)

# First plot: noisy signal
ax = fig.add_subplot(221)
ax.plot(t, hN, '-', c='gray')
ax.plot(t, np.zeros_like(t), ':k')
ax.text(0.98, 0.95, "Input Signal", ha='right', va='top',
        transform=ax.transAxes, bbox=dict(fc='w', ec='none'))

ax.set_xlim(0, 90)
ax.set_ylim(-0.5, 1.5)

ax.xaxis.set_major_locator(plt.MultipleLocator(20))
ax.set_xlabel(r'$\lambda$')
ax.set_ylabel('flux')

# Second plot: filtered signal
ax = plt.subplot(222)
ax.plot(t, np.zeros_like(t), ':k', lw=1)
ax.plot(t, h_smooth, '-k', lw=1.5, label='Wiener')
ax.plot(t, h_sg, '-', c='gray', lw=1, label='Savitzky-Golay')

ax.text(0.98, 0.95, "Filtered Signal", ha='right', va='top',
        transform=ax.transAxes)
ax.legend(loc='upper right', bbox_to_anchor=(0.98, 0.9), frameon=False)

ax.set_xlim(0, 90)
ax.set_ylim(-0.5, 1.5)

ax.yaxis.set_major_formatter(plt.NullFormatter())
ax.xaxis.set_major_locator(plt.MultipleLocator(20))
ax.set_xlabel(r'$\lambda$')

# Third plot: Input PSD
ax = fig.add_subplot(223)
ax.scatter(f[:N / 2], PSD[:N / 2], s=9, c='k', lw=0)
ax.plot(f[:N / 2], P_S[:N / 2], '-k')
ax.plot(f[:N / 2], P_N[:N / 2], '-k')

ax.plot(f[:N / 2], 3000*Phi[:N / 2], '-c')

ax.text(0.98, 0.95, "Input PSD", ha='right', va='top',
        transform=ax.transAxes)

ax.set_ylim(-100, 3500)
ax.set_xlim(0, 0.9)

ax.yaxis.set_major_locator(plt.MultipleLocator(1000))
ax.xaxis.set_major_locator(plt.MultipleLocator(0.2))
ax.set_xlabel('$f$')
ax.set_ylabel('$PSD(f)$')

# Fourth plot: Filtered PSD
ax = fig.add_subplot(224)
filtered_PSD = (Phi * abs(HN)) ** 2
ax.scatter(f[:N / 2], filtered_PSD[:N / 2], s=9, c='k', lw=0)

ax.text(0.98, 0.95, "Filtered PSD", ha='right', va='top',
        transform=ax.transAxes)

ax.set_ylim(-100, 3500)
ax.set_xlim(0, 0.9)

ax.yaxis.set_major_locator(plt.MultipleLocator(1000))
ax.yaxis.set_major_formatter(plt.NullFormatter())
ax.xaxis.set_major_locator(plt.MultipleLocator(0.2))
ax.set_xlabel('$f$')

plt.show()
```
