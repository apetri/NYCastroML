# Nonlinear regression

Nonlinear regression pertains to models that are __nonlinear in the parameters__ -- the model:

$$
    f(x) = a x^4 + b \sin{x}
$$

is still linear in the parameters $\boldsymbol{\theta} = (a,b)$. That is, if we measure some dependent variables $\boldsymbol{y}=y_i$, we can write the model as:

$$
    \boldsymbol{y} = \boldsymbol{X}\boldsymbol{\theta} + \boldsymbol{\epsilon}
$$

where $\boldsymbol{X}$ is a matrix of the independent variables and $\boldsymbol{\epsilon}$ is some error or noise. In the case above, if we measured two points $(x_1,y_1)$, $(x_2,y_2)$, this would look like:

$$
    \begin{pmatrix}y_1\\y_2\end{pmatrix} = \begin{pmatrix}x_1^4~~~\sin{x_1}\\x_2^4~~~\sin{x_2}\end{pmatrix}~\begin{pmatrix}a\\b\end{pmatrix}
$$

---

Models that are nonlinear in parameters are not unusual. For example, fitting an exponential function

$$
    f(t) = Ae^{\lambda t}
$$

is nonlinear in the parameter $\lambda$. However, in some cases these models can be transformed (via a coordinate transformation) to a model that is linear. Here, we could instead fit

$$
    g(t) = \log f(t) = \log A + \lambda t  
$$

which we could then write as

$$
    \begin{pmatrix}y_1\\y_2\\\vdots\\y_n\end{pmatrix} = \begin{pmatrix}1~~~t_1\\1~~~t_2\\\vdots\\1~~~t_n\end{pmatrix}~\begin{pmatrix}\log A\\\lambda\end{pmatrix}
$$

This distinction is only useful if (1) speed is critical or (2) you are getting bad performance/convergence fitting the nonlinear model.

---

There are a number of algorithms that can solve nonlinear regression problems. Depending on the method, they may require explicitly  supplying the Jacobian (and sometimes the Hessian) of the model with respect to the model parameters. For example, for the model above

$$
    f(t) = Ae^{\lambda t}
$$

the Jacobian would be

$$
    \boldsymbol{J} = \left( \begin{array}{cc}
    \frac{\partial f}{\partial A}(t_1) & \frac{\partial f}{\partial \lambda}(t_1) \\
    \frac{\partial f}{\partial A}(t_2) & \frac{\partial f}{\partial \lambda}(t_2) \\
    \vdots & \vdots \\
    \frac{\partial f}{\partial A}(t_n) & \frac{\partial f}{\partial \lambda}(t_n) \\
    \end{array} \right) = 
    \left( \begin{array}{cc}
    e^{\lambda t_1} & At_1 e^{\lambda t_1} \\
    e^{\lambda t_2} & At_2 e^{\lambda t_2} \\
    \vdots & \vdots \\
    e^{\lambda t_n} & At_n e^{\lambda t_n} \\
    \end{array} \right)
$$

Nonlinear methods that don't require a Jacobian will probably just numerically estimate it for you.

---

Scipy includes fast implementations of many algorithms, but we'll use the Levenberg-Marquardt algorithm through `scipy.optimize.leastsq`. Let's try fitting the above model to some fake data.

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import scipy.optimize as so
```

```python
def model(p, t):
    A,lambd = p
    return A*np.exp(lambd*t)

ndata = 25
true_p = (0.5, 0.2) # A, lambda

# generate the fake data
t = np.random.uniform(0., 10., size=ndata)
t.sort()

sigma_y = np.ones(ndata)*0.2
y = model(true_p, t) + np.random.normal(0.,sigma_y)
```

```python
# plot the "data"
plt.plot(t, y, marker='o', linestyle='none')

# plot the true line
_t = np.linspace(t.min(),t.max(),100)
plt.plot(_t, model(true_p,_t), marker=None, linestyle='-')
```

```python
def jacobian(p, t, y, sigma_y):
    A,lambd = p
    return np.array([np.exp(lambd*t), A*t*np.exp(lambd*t)])

def error_func(p, t, y, sigma_y):
    return (model(p, t) - y) / sigma_y
```

```python
fit_p,cov,infodict,mesg,ier = so.leastsq(error_func, [0.,0.], args=(t, y, sigma_y), 
                                         Dfun=jacobian, col_deriv=1, full_output=True)
```

```python
# plot the "data"
plt.plot(t, y, marker='o', linestyle='none')

# plot the true line
_t = np.linspace(t.min(),t.max(),100)
plt.plot(_t, model(true_p,_t), marker=None, linestyle='-')

# plot the fit line
plt.plot(_t, model(fit_p,_t), marker=None, linestyle='-')
```

```python
from matplotlib.patches import Ellipse

# plot error contours
U, S, V = np.linalg.svd(cov)
theta = np.degrees(np.arctan2(U[1, 0], U[0, 0]))

fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111)
for factor,alpha in zip([1,2,3],[0.7,0.5,0.3]):
    ellipse = Ellipse(xy=fit_p,
                      width=2*np.sqrt(S[0])*factor,
                      height=2*np.sqrt(S[1])*factor,
                      angle=theta, facecolor='#666666', 
                      edgecolor='#333333', alpha=alpha)
    ax.add_patch(ellipse)

# draw lines for true values
ax.axvline(true_p[0], alpha=0.75, linewidth=2., color='#9ecae1') # A
ax.axhline(true_p[1], alpha=0.75, linewidth=2., color='#9ecae1') # lambda
       
ax.set_xlim(0.,1.) # A
ax.set_ylim(0.,0.5) # lambda
```
