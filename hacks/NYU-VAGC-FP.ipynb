# The Fundamental Plane in the NYU-VAGC

## Introduction

We obtained our sample of galaxies from the [NYU Value-Added Galaxy Catalog](http://sdss.physics.nyu.edu/vagc/) (VAGC, Blanton et al. 2005a). This sample includes K corrections calculated with the methods of Blanton & Roweis (2007). K corrections are available for both Petrosian & Model magnitudes. At this exploratory stage, we have not established a preference for one over the other, but instead have created plots with both. However, to conserve space, we have simply chosen to display plots made with model magnitudes. The plots tend to be extremely similar, so nothing is lost by doing this.
The K corrections assume the following cosmology:

$$\Omega_{M} = 0.3$$
$$\Omega_{\Lambda} = 0.7$$
$$H_0 = 100 \mathrm{km} \, \mathrm{sec}^{-1} \, \mathrm{Mpc}^{-1}$$

We obtained the distance modulus by subtracting the K correction and absolute magnitude from the observed magnitude, that is

$$\mu = m - M - K$$

In all that follows, $R_e$ is the effective radius in kpc, $I_e$ is mean surface brightness within the effective radius in $L_{\odot} \mathrm{pc}^{-2}$ & $\sigma$ is the velocity dispersion in $\mathrm{km}\, \mathrm{sec}^{-1}$. Where magnitudes are used in plots, they are in the SDSS $r$ band.

## Data and Cuts

Cuts were applied to the VAGC sample in two stages. There was a first cut on general data quality, primarily to ensure that all quantities that appear in logarithms are positive definite. This cut also included the redshift bounds $0.02 < z < 0.08$. Velocity dispersion was required to be in the range $90 \mathrm{km}\, \mathrm{sec}^{-1} < \sigma < 400 \mathrm{km}\, \mathrm{sec}^{-1}$. See Bernardi et al. (2006) for a discussion of the problem of superposed objects creating large apparent velocity dispersions. The number of expected galaxies given mean density as defined in the VAGC 'simpleden' density estimator was required to be greater than 0.35. Finally, the VAGC Main sample bit was required to be set.

A second cut was applied to select elliptical galaxies from the sample created by the first cut. We emphasize that this cut is purely empirical, although similar cuts have been used by other authors. There is some degeneracy among these cuts, & it should be possible to do much better, for example by including spectroscopic information, but this cut suffices for initial experiments. The second-round cuts are as follows:

* $\sigma_{\sigma} < 0.1 \sigma$ (that is, velocity dispersion error less than 10%)
* Sérsic index > 3.0 (from VAGC Sérsic fits)
* $R_{90}/R_{50} > 2.5$ (we use Petrosian radii for this measure of concentration)
* FRACDEV > 0.8 (weight of de Vaucouleurs component in best composite model)
* $r_{\mathrm{PSF}} – r_{\mathrm{MODEL}} > 0.3$  (star-galaxy separation, Strauss et al. 2002)
* $g – r > -0.03 (M_r + 2)$ (this is a 'red sequence' cut)

```python
#
# Read data
#
# If the parameters don't change, this reads from a pickle file instead of directly from the VAGC.
#
d = load_vagc(rows,options.imagedir,options.minfrac,clobber=options.clobber)
```

The cuts described above remove a very large number of objects from consideration.  Typically only about 2% pass the first cut, & 1% pass the second cut.

## 'Traditional' PCA Analysis

There are many examples in the literature where the Fundamental Plane is characterized by finding the plane that minimizes the variance of the input data.  This is equivalent to a simple PCA transformation.  This method can be found *e.g.* in Binney & Merrifield (1998).  I have updated the notation & method to reflect Section 7.3.1 of the astroML book. 

We construct a matrix $X$ whose ($K = 3$) rows are the (common) log of the effective radius, the log of the surface brightness and the log of the velocity dispersion, with the mean subtracted from each column.

$$X = \left\{ \log_{10} R_e, \log_{10} \sigma, \log_{10} I_e \right\} -  \left\{ \left\langle \log_{10} R_e \right\rangle, \left\langle \log_{10} \sigma \right\rangle , \left\langle \log_{10} I_e \right\rangle \right\}$$

Then we construct the covariance matrix $C_X$ by multiplying

$$C_X = \frac{1}{N-1} X^T \cdot X $$

The PCA step reduces to finding the eigenvectors of $C_X$.  We'll use `numpy.linalg.eigh()` for this.  Once we have the eigenvectors, we order them by corresponding eigenvalue and select the eigenvector with the smallest eigenvalue.  This vector is called $R_{\mathrm{min}}$.  We want to convert this into the equation of a plane.  Since we have subtracted the mean, we should be able to set

$$Y_{\mathrm{min}} = X \cdot R_{\mathrm{min}} = 0$$

The 'canonical' form of the Fundamental plane is:

$$ \log_{10} R_e = a \log_{10} \sigma + b \log_{10} I_e + c $$

It is simple to show that

$$ a = - \frac{R_{\mathrm{min}}[1]}{R_{\mathrm{min}}[0]} $$

$$ b = - \frac{R_{\mathrm{min}}[2]}{R_{\mathrm{min}}[0]} $$

$$ c = \frac{\bar{X} \cdot R_{\mathrm{min}}}{R_{\mathrm{min}}[0]}$$

where $\bar{X}$ is the vector of mean values that was subtracted off to create the matrix $X$.

As you may have noticed, this analysis does not include errors in the data values in any way. It is merely is useful to get a very quick estimate of the fundamental plane parameters in the VAGC data. Bernardi et al. (2003) has an extensive discussion of the error analysis of fundamental plane fits.

```python
#
# Do traditional PCA
#
N = len(d[f]['Re'][w2,r])
K = 3
x = np.array([np.log10(d[f]['Re'][w2,r]),np.log10(d['spectro']['VDISP'][w2]),np.log10(d[f]['Ie'][w2,r])])
xbar = x.mean(axis=1)
X = x.transpose() - xbar
C_X = np.dot(X.transpose(),X)/float(N-1)
# print('X=',X)
# print('C_X=',C_X)
l,n = eigh(C_X)
eigsort = l.argsort()[::-1]
lsort = l[eigsort]
R = n[:,eigsort]
# print('l=',lsort)
# print('R=',R)
C_Y = np.dot(R.transpose(),np.dot(C_X,R))
# print('C_Y=',C_Y)
R_min = R[:,K-1]
# print('R_min=',R_min)
Y = np.dot(X, R)
# print('Y=',Y)
residual2 = Y[:,K-1]
# C = np.sum(residual2)/N
# print('C=',C)
# residual2 -= C
# print('residual2=',residual2)
x1 = np.array([np.log10(d[f]['Re'][w1,r]),np.log10(d['spectro']['VDISP'][w1]),np.log10(d[f]['Ie'][w1,r])])
x1bar = x1.mean(axis=1)
X1 = x1.transpose() - x1bar
Y1 = np.dot(X1,R)
residual1 = Y1[:,K-1]
# print('residual1=',residual1)
#
# log R = a log sigma + b log I + c
#
fp_a = -R_min[1]/R_min[0]
fp_b = -R_min[2]/R_min[0]
fp_c = np.dot(R_min,xbar)/R_min[0]
print(r"$$\log_{{10}} R_e = {0:+4.2f} \log_{{10}} \sigma {1:+4.2f} \log_{{10}} I_e {2:+4.2f}$$".format(fp_a,fp_b,fp_c))
#
# When we project so that R_e is on the x axis, this is the y axis.
#
Is1 = fp_b*np.log10(d[f]['Ie'][w1,r]) + fp_a*np.log10(d['spectro']['VDISP'][w1])
Is2 = fp_b*np.log10(d[f]['Ie'][w2,r]) + fp_a*np.log10(d['spectro']['VDISP'][w2])
```

So our Fundamental Plane equation is

$$\log_{10} R_e = +0.93 \log_{10} \sigma -0.76 \log_{10} I_e +0.18$$


```python
#
# The money plot
#
fig = plt.figure(dpi=100)
ax = fig.add_subplot(111)
p = ax.semilogx(d[f]['Re'][w1,r],Is1,'k.',
    d[f]['Re'][w2,r],Is2,'ro',
    [0.1, 100],[-1-fp_c, 2-fp_c],linecolour)
p[1].set_markerfacecolor('none')
p[1].set_markeredgecolor(squarecolour)
p[0].set_label('_nolegend_')
p[1].set_label(cutlabel)
p[2].set_label(r"$y = \log_{{10}} R_e {0:+4.2f}$".format(-(fp_c)))
ax.set_xlabel('$R_e$ [kpc]',fontproperties=titlefont)
ax.set_ylabel(r"${0:+4.2f} \log_{{10}} \sigma {1:+4.2f} \log_{{10}} I_e$".format(fp_a,fp_b),fontproperties=titlefont)
ax.set_title("NYU-VAGC IS-Radius Relation [{0} Fluxes]".format(d[f]['name']),fontproperties=titlefont)
ax.set_xlim(0.1,100)
ax.set_ylim(-1,4)
leg = ax.legend(loc=4,prop=legendfont,numpoints=1)
# fig.savefig(join(options.imagedir,"vagc_r-is_{0}.{1}".format(f,figext)))
# plt.close(fig)
```
